# d.Creating two 2-D arrays and Stacking them together both Horizontally and Vertically 

```python
#Defining first 2-D array
my_mat = np.array([[1, 2, 3], [ 4, 5, 6]])
print(my_mat)
```

```python
#Defining second 2-D array
my_matrix = np.array([[7, 8, 9], [10, 11, 12]])
print(my_matrix)
```

```python
#Horizontally stacking arrays together
np.hstack([my_mat, my_matrix])
```

```python
#Vertically stacking arrays together
np.vstack([my_mat, my_matrix])
```

## i.Concatenating data frames using dictionary

```python
#Defining dictinaries
GhanaCars = {'Company': ['Kantaka', 'VW', 'BMW', 'Neoplan'], 'Price': [23845, 171995, 135925, 71400]}
JapaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi'], 'Price': [29995, 23600, 61500, 58900]}

#Creating a DataFrame from the dictionaries
Ghacars = pd.DataFrame.from_dict(GhanaCars)
Japcars = pd.DataFrame.from_dict(JapaneseCars)

#Concatenating the two DataFrames
carsdf_concat = pd.concat([Ghacars, Japcars], keys=['Ghana', 'Japan'])
carsdf_concat
```
